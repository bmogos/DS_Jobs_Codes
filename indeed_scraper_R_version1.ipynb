## Extract the data for all the webpages according to the search criteria

```r
URL <- "https://ca.indeed.com/jobs?q=AWS&l=Canada&fromage=7"
webpage <-read_html(URL)

# #Get the number of jobs and number of pages
jobCount <- webpage %>% 
   rvest::html_nodes('body') %>% 
   xml2::xml_find_all("//div[contains(@id, 'searchCountPages')]") %>% 
   rvest::html_text()
   jobCount <- str_trim(stringr::str_replace_all(jobCount, "[\n]" , ""))
   jobCount <- unlist(stringr::str_split(jobCount, "of"))
   jobCount <- str_trim(jobCount[2])
   jobCount <- unlist(str_split(jobCount, " "))
   jobCount <- str_replace_all(jobCount, ",", "")
   jobCount <- as.integer(jobCount[1])
#jobCount

# #Get the number of pages
endPage <- floor(jobCount/15)*10 - 10

#Create an empty dataframe
job_postings <- data.frame() #create an empty dataframe

for(pageNum in seq(from=0, to=endPage, by=10)){
    url <- paste("https://ca.indeed.com/jobs?q=AWS&l=Canada&fromage=7&start=", page_number, sep="")
    #Get the source code from the given url
    webpage <-read_html(url)
    #webpage
    #str(webpage)
    
    #Get all the job instances: there are 15 job instances in the first page
    jobInstances <- html_nodes(webpage, ".jobsearch-SerpJobCard" ) 
    #jobInstances
    
    #Get all the nodes of the html <body>

    #body_nodes <- webpage %>% html_nodes('body')  #Same as body_nodes <- html_nodes(webpage, 'body')
    #body_nodes
    
    # 1. Get the jobId
    JobId <- jobInstances %>% html_attr("data-jk")
    
    # 2. Extract the job title
    title_html <- html_nodes(jobInstances, '.title a')
    title_text = html_text(title_html)
    JobTitle <- str_replace_all(title_text, "[\r\n]" , "") #replace the \r, \n by ""

    # 3. Get the names of the employer companies
    Company <- jobInstances %>% 
      rvest::html_node('span.company') %>% 
      rvest::html_text()
    Company <- stringr::str_replace_all(Company, "\n" , "") 
    
    # 4. Get all the locations
    Location <- jobInstances %>% html_nodes("div.recJobLoc") %>% html_attr("data-rc-loc")
    

    # 5. Get the summary of all the job instances
    Summary <- webpage %>% 
      rvest::html_nodes('body') %>% 
      xml2::xml_find_all("//div[contains(@class, 'summary')]") %>% 
      rvest::html_text()
    Summary <- str_replace_all(Summary, "\n" , "")
    
    # 6. Get the date the job was posted
    PostDate <- jobInstances %>% html_nodes("span.date") %>% html_text()
    
    # 7. Get the Collection date of the data
    CollectDate <- as.Date(character(0))  #This command creates an empty vector of type "Date"
    for(i in seq(from=1, to=length(JobTitle), by=1)){
        today <- Sys.Date()
        CollectDate <- list.append(CollectDate, today)
    }

    
    # 8. Get the salary of all the job instances
    Salary <- jobInstances %>% 
      rvest::html_node('span.salaryText') %>%
      rvest::html_text()
    Salary <- str_replace_all(Salary, "\n" , "")
    
    # 9. Get the link of each job posting
    JobUrl <- html_nodes(jobInstances, '.title a') %>% html_attr('href') %>% paste('https://ca.indeed.com', . , sep='') 
    
    
    # 10. Get the job rating
    JobRating <- jobInstances %>% 
      rvest::html_node('span.ratingsContent') %>%
      rvest::html_text()
    JobRating <- str_replace_all(JobRating, "\n" , "")
    
    
    
    job_postings <- rbind(job_postings, data.frame(JobId, JobTitle, Company, Location, Salary, PostDate, JobUrl, CollectDate, Summary, JobRating))
}

```

```r
# 11. Get the job description
Description <- vector(mode = "list")
for(i in seq(from=1, to=dim(job_postings)[1], by=1)){
    link <- toString(job_postings$JobUrl[1])
    webpageCard <-read_html(link)

    discr <- webpageCard %>% 
    rvest::html_nodes('div.jobsearch-jobDescriptionText') %>%
    rvest::html_text()
    discr <- str_replace_all(discr, "\n" , "")
    Description <- list.append(Description, discr)
    Description <- as.character(Description)  
}
```

```r
#Add the Description column into the dataframe
for(i in seq(from=1, to=dim(job_postings)[1], by=1)){
    job_postings$Description[i] <- Description[i]
}

dim(job_postings)
```
